```python
#Got Success...
#Create a small python program using import mysql.connector and import Tkinder as well as connect with MySQL database. Ask the two inputs from the user and select  from radio button which operation has to done (ADDITION, SUBTRACTION, MULTIPLICATION, DIVISION). All the records of operations have to store in MySQL database like First_Number,  Second_Number, Operation, Result



import mysql.connector
from tkinter import*
#import tkinter as tk

#Connecting to MySQL database...
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="calculator"
)

#Creating a cursor object...
cursor = db.cursor()

#Creating a Tkinter window...
root = Tk()
root.title("Calculator")
#root.geometry("550x165")

#Function to perform addition...
def add():
    first_number = float(num1.get())
    second_number = float(num2.get())
    result = first_number + second_number
    result_var.set(result)
    save_operation(first_number, second_number, "ADDITION", result)

#Function to perform subtraction...
def subtract():
    first_number = float(num1.get())
    second_number = float(num2.get())
    result = first_number - second_number
    result_var.set(result)
    save_operation(first_number, second_number, "SUBTRACTION", result)

#Function to perform multiplication...
def multiply():
    first_number = float(num1.get())
    second_number = float(num2.get())
    result = first_number * second_number
    result_var.set(result)
    save_operation(first_number, second_number, "MULTIPLICATION", result)

#Function to perform division...
def divide():
    first_number = float(num1.get())
    second_number = float(num2.get())
    if second_number != 0:
        result = first_number / second_number
        result_var.set(result)
        save_operation(first_number, second_number, "DIVISION", result)
    else:
        result_var.set("Error: Division by zero")

#Function to save operation details to the database...
def save_operation(first_number, second_number, operation, result):
    cursor.execute("INSERT INTO Operations (First_Number, Second_Number, Operation, Result) VALUES (%s, %s, %s, %s)",(first_number, second_number, operation, result))
    db.commit()
    
#Function to reset all fields...
def reset():
    num1.delete(0, END)
    num2.delete(0, END)
    result_var.set("")

#Labels...
Label(root, text="Enter first number:").grid(row=0, column=0)
Label(root, text="Enter second number:").grid(row=1, column=0)
Label(root, text="Select Operation:").grid(row=2, column=0)
Label(root, text="Result:").grid(row=3, column=0)

#Entry fields...
num1 = Entry(root)
num1.grid(row=0, column=1)
num2 = Entry(root)
num2.grid(row=1, column=1)

#Result field...
result_var = StringVar()
result_field = Entry(root, textvariable=result_var, state='disabled')
result_field.grid(row=3, column=1)

#Radio buttons for operations...
operation_var = StringVar()
operation_var.set("ADDITION")
add_radio = Radiobutton(root, text="Addition", variable=operation_var, value="ADDITION")
add_radio.grid(row=2, column=1)
subtract_radio = Radiobutton(root, text="Subtraction", variable=operation_var, value="SUBTRACTION")
subtract_radio.grid(row=2, column=2)
multiply_radio = Radiobutton(root, text="Multiplication", variable=operation_var, value="MULTIPLICATION")
multiply_radio.grid(row=2, column=3)
divide_radio = Radiobutton(root, text="Division", variable=operation_var, value="DIVISION")
divide_radio.grid(row=2, column=4)

#Button to calculate result...
Button(root, text="Calculate", command=lambda: perform_operation(operation_var.get())).grid(row=3, column=1, columnspan=4)
Button(root, text="Reset", command=reset).grid(row=4, column=0, columnspan=3)

#Function to perform selected operation...
def perform_operation(operation):
    if operation == "ADDITION":
        add()
    elif operation == "SUBTRACTION":
        subtract()
    elif operation == "MULTIPLICATION":
        multiply()
    elif operation == "DIVISION":
        divide()

root.mainloop()

# Close cursor and database connection
cursor.close()
db.close()
```
